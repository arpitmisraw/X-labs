```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('USvideos.csv')
views = df['views'].values.reshape(15593, 1)
likes = df['likes'].values.reshape(15593, 1)
```

```python
views_train = views[2000: 3000]
views_test = views[: 20]
```

```python
likes_train = likes[2000: 3000]
likes_test = likes[: 20]
```

```python
rgm = linear_model.LinearRegression()
rgm.fit(views_train, likes_train)
```

```python
plt.scatter(views_test, likes_test, color = 'b')
plt.plot(views_test, rgm.predict(views_test), color = 'r')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Estimation of  number of Likes from number of Views')
plt.savefig('youtube_plot')
plt.show()
```
